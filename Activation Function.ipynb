```python
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

### Linear activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def linear(number):
    return number

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
plt.plot(number, linear(number))
plt.plot([0.5])
plt.title("Linear Activation Function")
plt.show()
```

### Binary Step Activation Function
### Binary step function returns value either 0 or 1.

It returns '0' if the input is the less then zero

It returns '1' if the input is greater than zero

Link: https://numpy.org/doc/stable/reference/generated/numpy.heaviside.html

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def binaryStep(number):
    return np.heaviside(number, 1)

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
plt.plot(number, binaryStep(number))
plt.plot([0.5])
plt.title("Binary Step Activation Function")
plt.show()
```

### Sigmoid activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def sigmoid(number):
    return 1 / (1 + np.exp(-number)) # We can also write math.exp(-number)

number = np.linspace(-10, 10)
print("The numbers are: ", number)
plt.plot(number, sigmoid(number))
plt.plot([0.5])
plt.title("Binary Sigmoid Activation Function")
plt.show()
```

### Tanh activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def tanh(number):
    return np.tanh(number)

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
plt.plot(number, tanh(number))
plt.title("Hyperbolic Tan Activation Function")
plt.show()
```

### ReLU(Rectified Linear Unit) activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def ReLU(number):
    numberStore = []
    for i in number:
        initial_value = 0
        if i >=0:
            initial_value = i
        numberStore.append(initial_value)
    return numberStore

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
result = ReLU(number)
print(result)
plt.plot(number, ReLU(number))
plt.title("ReLU Activation Function")
plt.show()
```

### Leaky ReLU activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def leakyReLU(number):
    numberStore = []
    for i in number:
        if i >=0:
            numberStore.append(i)
        else:
            numberStore.append(0.01 * i)
    return numberStore

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
result = leakyReLU(number)
print(result)
plt.plot(number, leakyReLU(number))
plt.title("Leaky ReLU Activation Function")
plt.show()
```

### ELU (Exponential Linear Unit) activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def elu(number):
    numberStore = []
    for i in number:
        if i>=0:
            numberStore.append(i)
        else:
            numberStore.append(0.5 * (np.exp(i) - 1)) # here 0.5 could be change not constant
    return numberStore

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
result = elu(number)
print(result)
plt.plot(number, elu(number))
plt.title("ELU Activation Function")
plt.show()
```

### SoftMax activation function

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,5))

def softmax(number):
    return np.exp(number) / np.sum(np.exp(number))

number = np.linspace(-10, 10)
# print("The numbers are: ", number)
result = softmax(number)
print(result)
plt.plot(number, softmax(number))
plt.title("Softmax Activation Function")
plt.show()
```
